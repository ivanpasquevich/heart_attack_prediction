```python
# levanto el dataframe y lo guardo en la variable data
data = pd.read_csv("..\Data\heart.csv")
```

```python
# creo columna de nulos
data["nulos"] = np.nan
```

```python
# relleno nulos con valor random entre 9 y 99
data["nulos"] = data["nulos"].fillna(random.randint(9,99))
```

```python
data
```

```python
# descarto la columna de nulos
data = data.drop("nulos",axis=1)
```

### Visualizacion


```python
# grafico las ocurrencias del campo target
plt.figure(figsize=(7,5))
ax = sns.countplot(x="target",data=data,palette='rocket')
```

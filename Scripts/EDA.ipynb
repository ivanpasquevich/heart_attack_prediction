```python
# creamos dataframe con la modificacion esa (variable nominal thal)
data_modif = data.copy()
```

```python
def categorizacion(val):
    if val==0:
        return "normal"
    elif val==1:
        return "fixed defect"
    elif val==2:
        return "reversable defect"
    else:
        return "unknown"
```

```python
data_modif["thal_cat"] = data_modif["thal"].apply(categorizacion)
data_modif[["thal","thal_cat"]]
```

```python
data_modif.drop('thal',axis=1,inplace=True)
data_modif
```

```python
dum = pd.get_dummies(data_modif['thal_cat'])
data_modif = pd.concat([data_modif, dum], axis=1) 
data_modif
```

```python
data_modif.drop('thal_cat',axis=1,inplace=True)
data_modif
```

### Modelo de machine learning


```python
# random forest
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)

#train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.drop(['target'],axis=1), 
                                                    data['target'], test_size=0.20, 
                                                    random_state=101)

rfc.fit(X_train, y_train)

rfc_pred = rfc.predict(X_test)

from sklearn.metrics import classification_report,confusion_matrix

print(classification_report(y_test,rfc_pred))
print(confusion_matrix(y_test,rfc_pred))

```
